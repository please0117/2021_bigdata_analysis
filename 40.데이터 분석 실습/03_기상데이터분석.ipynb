```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('seoul_temperature.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('지점', axis=1, inplace=True)
df.head()
```

```python
df.columns = ['날짜', '평균', '최저', '최고']
df.head()
```

```python
df.날짜 = pd.to_datetime(df.날짜)
df.info()
```

```python
df['년도'] = df.날짜.dt.year
df['월'] = df.날짜.dt.month
df['일'] = df.날짜.dt.day
df.head()
```

```python
# 평균 기온이 상승했을까? (8월 1일)
df_8 = df[(df.월==8) & (df.일==1)][['년도', '평균']]
df_8.head()
```

```python
# 1월 1일
df_1 = df[(df.월==1) & (df.일==1)][['년도', '평균']]
df_1.head()
```

```python
# 마이너스 기호 

plt.figure(figsize=(10, 5))
sns.lineplot(x='년도', y='평균', data=df_8)
sns.lineplot(x='년도', y='평균', data=df_1)
plt.show()
```

```python
# 가장 더웠던 날과 최고 기온은?
df_max = df.sort_values(by='최고', ascending=False)
df_max.head(10)
```

```python
# 가장 추웠던 날과 최저 기온은?
df_min = df.sort_values(by='최저')
df_min.head(10)
```

```python
# 일교차가 가장 큰 달은?
df['일교차'] = df.최고 - df.최저
df.head()
```

```python
pd.DataFrame(df.groupby('월')['일교차'].mean())
```
