# 4. Bar Plot
* 막대 그래프로 여러 값들을 비교할 때 유용함
* 데이터
    * 수치형 데이터 / 양적 데이터(numerical data) : 키, 몸무게, 시력, 머리카락 개수
    * 범주형 대이터(categorical data) : 혈액형, 성별, 지역
* x축에는 범주형 데이터, y축에는 수치형 데이터

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 년도별 출생지수 비교
years = ['2018', '2019', '2020']
number = [100, 90, 80]

# bar(x, height, ...)
plt.bar(years, number, width=0.5)
plt.title('년도별 출생지수', size=20)
plt.xlabel('년도')
plt.ylabel('출생지수')
plt.show()
```

```python
# 수평 막대그래프 그리기
# barh(y, width)
plt.barh(years, number, height=0.5)
plt.title('년도별 출생지수', size=20)
plt.xlabel('출생자수')
plt.ylabel('년도')
plt.show()
```

```python
# 년도, 성별로 출생자수 비교하가
male = [100, 90, 80]
female = [90, 80, 70]
x = np.arange(3)  # 0~2

# bar(x, height)
plt.bar(x, male, width=0.5)
plt.bar(x, female, width=0.5)

plt.xticks(x, years)
plt.show()
```

```python
bar_width = 0.3
plt.bar(x - bar_width/2, male, width=bar_width)
plt.bar(x + bar_width/2, female, width=bar_width)
plt.xticks(x, years)
plt.legend(['male', 'female'])
plt.show()
```
