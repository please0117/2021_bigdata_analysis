# 5. Pie Chart
* 카데고리별 또는 그룹별로 집계를 하고 백분율로 표시

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [1, 2, 3, 4]
plt.pie(data)
plt.show()
```

```python
labels = ['순대', '치킨', '초밥', '떡볶이']
plt.pie(data, labels=labels)
plt.show()
```

```python
# 퍼센트 출력하기
plt.pie(data, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
explode = [0.2, 0, 0, 0]
colors = ['r', 'g', 'b', 'y']
plt.pie(data, labels=labels, colors=colors, explode=explode, shadow=True)
plt.show()
```
